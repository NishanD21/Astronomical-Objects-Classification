## Impact of redshift for the performance of each class

The best performimg model was identified as Random Forest and the best performing normlization method was identifed as the Z-Score method. Therefore, the performance for each class was determined using the Random forest algorithm and the dataset normalized using Z-Score method.

```python
import pandas as pd # for Dataframe manupulation
import matplotlib.pyplot as plt #for data visualization
import seaborn as sns #for data visualization
import numpy as np # for arrays manipulation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc 
# to import evaluation metrices
```

#### Dataset without redshift

```python
df_zc_wr = pd.read_csv("D:/Academic/MSc_Data_Science/Course/8.Technology Dissertation/4. Dissertation/Datasets/sgq_classification_zc_wr.csv")
```

```python
df_zc_wr
```

```python
x = df_zc_wr.drop("class",axis='columns')
x.head()
```

```python
y = df_zc_wr["class"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier

# initializing model
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(x_train,y_train)
```

```python
y_pred = model_rf.predict(x_test)
```

```python
# Evaluating the model's performance using accuracy, precision, recall and F1 Score for each class separately.
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average=None)
recall = recall_score(y_test, y_pred, average=None)
f1 = f1_score(y_test, y_pred, average=None)

# calculating the accuracy for each class separately.
class_accuracy = {}
for i in range(len(precision)):
    tp = sum((y_test == i) & (y_pred == i))
    tn = sum((y_test != i) & (y_pred != i))
    fp = sum((y_test != i) & (y_pred == i))
    fn = sum((y_test == i) & (y_pred != i))
    class_accuracy[i] = (tp + tn) / (tp + tn + fp + fn)

# printing the evaluation results for each class separately.
for i in range(len(precision)):
    print("Evaluation results for class", i)
    print("Accuracy:", class_accuracy[i])
    print("Precision:", precision[i])
    print("Recall:", recall[i])
    print("F1 Score:", f1[i])
    print()
```
